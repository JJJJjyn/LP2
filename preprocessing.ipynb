```python
import os 
import re
from os import path 
from bs4 import BeautifulSoup
from simpletransformers.language_representation import RepresentationModel
```

```python
def scaner_file (url):
    file  = os.listdir(url)
    file_name = []
    for f in file:
        real_url = path.join (url , f)
        file_name.append(path.abspath(real_url))
            # if path.isfile(real_url):
                    # print(path.abspath(real_url))
    return file_name
```

```python
def read_xml(url):
    with open(url, 'rb') as f:
        data = f.read()
    bs_data = BeautifulSoup(data, 'xml') 
    b_document = bs_data.find_all('document') 
    return b_document
```

```python
import xml.etree.ElementTree as ET

def read_xml2(url):
    tree = ET.parse(url)
    root = tree.getroot()
    b_document = []
    for documents in root:
        for document in documents:
            b_document.append(document.text)
    return b_document
```

```python
pattern = r"(<[A-Za-z0-9]+>)|(</[A-Za-z0-9]+>)|(#[A-Za-z0-9]+#)"
```

```python
pattern = r"(<[A-Za-z0-9]+>)|(</[A-Za-z0-9]+>)|(#[A-Za-z0-9]+#)"
model = RepresentationModel(
        model_type="bert",
        model_name="bert-base-uncased",
        use_cuda=False
    )
```

```python
def norm(document):
    norm_doc = []
    for sentence in document:
        norm_sent = ' '.join(re.sub(pattern," ",str(sentence)).split())
        norm_doc.append(norm_sent)
    return norm_doc
```
